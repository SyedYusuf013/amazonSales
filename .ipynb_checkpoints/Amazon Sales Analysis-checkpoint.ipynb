## Exploratory Data Analysis (EDA)

### Visualizing Rating Distribution

```python
plt.figure(figsize = (10,8))
vf = sns.countplot(x='rating', data = df)
vf.bar_label(vf.containers[0], fontsize = 14)

plt.title("Distribution of Rating", fontsize = 14)
plt.xlabel("Rating", fontsize=14, labelpad=15)  # Increase x-axis label size
plt.ylabel("Count", fontsize=14, labelpad=15)  # Increase y-axis label size
plt.xticks(fontsize=14)  # Increase x-tick labels size
plt.yticks(fontsize=14) # Increase y-tick labels size

plt.show()
```

### Analyzing Yearly Sales

```python
plt.figure(figsize=(16, 8))

# Extracting year from timestamp
df['year'] = pd.DatetimeIndex(df['timestamp']).year

# Grouping by year and counting the ratings (which represents sales)
sales_per_year = df.groupby('year')['rating'].count()

# Plotting the bar graph
ax = sales_per_year.plot(kind='bar')

# Adding title and labels
plt.title("Best Year for Sales", fontsize=14, pad=10)
plt.xlabel("Year", fontsize=14, labelpad=15)  # Increase x-axis label size
plt.ylabel("Sales", fontsize=14, labelpad=15)  # Increase y-axis label size
plt.xticks(fontsize=14)  # Increase x-tick labels size
plt.yticks(fontsize=14)  # Increase y-tick labels size

# Loop through the bars and add the sales count on top of each bar
for i, value in enumerate(sales_per_year):
    ax.text(i, value + 5000, f'{value}', ha='center', fontsize=14)  # Adjust '5000' for the text position

plt.show()
```

### The year 2015 had the highest number of sales.

### Brand Sales in Year 2015

### Brand Sales in Rating

```python
# Get the top 10 brands based on count
top_brands = df['brand'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 brands
filtered_df = df[df['brand'].isin(top_brands)]

# Create the count plot
plt.figure(figsize=(30, 15))
ax = sns.countplot(x="brand", data=filtered_df, hue="rating", order=top_brands)

# Add bar labels for each container
for container in ax.containers:
    ax.bar_label(container, fontsize=20)
    
# Adding title and labels
plt.title("ratings of the brand", fontsize=20, pad=10)
plt.xlabel("brand", fontsize=20, labelpad=15)  # Increase x-axis label size
plt.ylabel("Sales", fontsize=20, labelpad=15)  # Increase y-axis label size
plt.xticks(fontsize=20)  # Increase x-tick labels size
plt.yticks(fontsize=20)  # Increase y-tick labels size

plt.legend(title="rating", fontsize=18, title_fontsize=20)

plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.show()
```

### The visualization shows how the top 10 brands are rated by customers, with distinct rating distributions across brands, indicating varying levels of customer satisfaction.

### Brand Sales by last 5 Years

```python
# Get the top 10 brands based on count
top_brands = df['brand'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 brands and the last 5 years
last_5_years = df['year'].max() - 5
filtered_df = df[(df['brand'].isin(top_brands)) & (df['year'] > last_5_years)]

# Create the count plot with 'rating' and 'year' as hue
plt.figure(figsize=(30, 15))
ax = sns.countplot(x="brand", data=filtered_df, hue="year", order=top_brands)

# Add bar labels for each container
for container in ax.containers:
    ax.bar_label(container, fontsize=14)

# Adding title and labels
plt.title("Brand Sales by Year (Last 5 Years)", fontsize=22, pad=10)
plt.xlabel("Brand", fontsize=20, labelpad=15)  # Increase x-axis label size
plt.ylabel("Sales", fontsize=20, labelpad=15)  # Increase y-axis label size
plt.xticks(fontsize=20)  # Increase x-tick labels size
plt.yticks(fontsize=20)  # Increase y-tick labels size

# Adjust the legend to show both year and rating
plt.legend(title="Year", fontsize=18, title_fontsize=20)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

### The visualization reveals that sales for the top 10 brands have fluctuated over the last 5 years, with some brands experiencing more consistent performance across multiple years.

### Brand Sales by rating (Multiple Year Subplots)

```python
# Filter the DataFrame to include only the last 6 years
last_6_years = df['year'].max() - 6
filtered_df = df[df['year'] > last_6_years]

# Get the top 10 brands based on their counts across the entire filtered dataset
top_10_brands = filtered_df['brand'].value_counts().head(10).index

# Further filter the DataFrame to include only the top 10 brands
filtered_df = filtered_df[filtered_df['brand'].isin(top_10_brands)]

# Get unique years from the filtered data
years = sorted(filtered_df['year'].unique())

# Number of columns and rows for the subplot grid
n_cols = 2
n_rows = 3  # Fixed to 3 rows for last 6 years

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 5))  # Adjust figsize as needed

# Flatten the axes array for easy iteration (handles both 1D and 2D arrays)
axes = axes.flatten()

# Iterate over years and plot count plots for each year
for i, year in enumerate(years):
    year_df = filtered_df[filtered_df['year'] == year]
    ax = sns.countplot(x='brand', data=year_df, hue='rating', ax=axes[i], order=top_10_brands)
    
    # Add bar labels for each container
    for container in ax.containers:
        ax.bar_label(container)
        
    axes[i].set_title(f'Brand Sales by Rating in {year}')
    axes[i].set_xlabel('Brand')
    axes[i].set_ylabel('Count')
    plt.xticks(rotation=45)

# Remove empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

### The visualization highlights the variation in sales and ratings across the top 10 brands over the last 6 years, with noticeable brand-specific trends in performance each year.
